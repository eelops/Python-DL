### 综合题 1：Softmax（一维）

**定义**：对向量 $\mathbf{z} \in \mathbb{R}^d$，其 softmax 的第 $i$ 个分量为

$$\mathrm{softmax}(\mathbf{z})_i = \frac{\exp(z_i)}{\sum_{j=1}^{d} \exp(z_j)}.$$

**直观**：先把每个分量取指数，再除以「全体指数之和」，因此结果**非负**，且**所有分量之和为 1**，可看成在 $d$ 个类别上的概率分布。

**数值稳定**：指数容易溢出。常用技巧是先减去 $m = \max_k z_k$（对全体分量相同），即计算 $\mathrm{softmax}(\mathbf{z}) = \mathrm{softmax}(\mathbf{z} - m\mathbf{1})$，其中 $\mathbf{1}$ 为全 1 向量。因为 softmax 对「整体加常数」不变，这样 $\exp(z_i-m)$ 最大不超过 1。

**在代码里**：对应「按元素指数」「对分量求和得到标量」「用广播把和除到每个分量上」。请**不要**直接调用现成的 softmax 完成本题主体（最后可自行用库函数核对）。


```python
z = torch.tensor([2.0, 1.0, 0.1, -5.0], dtype=torch.float32)

# 请在此实现：数值稳定的 softmax(z)，结果命名为 p，形状与 z 相同




print("p:", p)
if p is not None:
    print("分量之和（应接近 1）:", p.sum())
# 如需核对，可用框架自带的 softmax 在同一输入上对比（请自行查阅文档）
```
